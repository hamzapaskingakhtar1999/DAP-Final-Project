# Log transformation for handling skewness

```python
import numpy as np
# Assuming 'joined_data' is a DataFrame object in pandas
selected_columns = ['budget', 'vote_count', 'vote_average', 'popularity']
selected_data = joined_data[selected_columns]

# Apply log transformation to selected columns
log_transformed_data = selected_data.apply(lambda x: np.log(x + 1))  # Adding 1 to avoid log(0)

# Display the log-transformed data
print(log_transformed_data)
```

# Check for outliers

```python
# Assuming 'joined_data' is a DataFrame object in pandas
selected_columns = ['budget', 'vote_count', 'vote_average', 'popularity']
selected_data = log_transformed_data[selected_columns]

# Calculate z-scores for each selected column
z_scores = (selected_data - selected_data.mean()) / selected_data.std()

# Define threshold for outlier detection (e.g., z-score > 3)
threshold = 3

# Identify outliers
outliers = (z_scores > threshold) | (z_scores < -threshold)

# Display outliers
print(outliers)
```

```python
import pandas as pd


selected_columns = [ 'budget', 'vote_count', 'vote_average', 'popularity']
selected_data = log_transformed_data[selected_columns]

# Calculate quartiles
Q1 = selected_data.quantile(0.10)
Q3 = selected_data.quantile(0.90)

# Calculate IQR (Interquartile Range)
IQR = Q3 - Q1

# Define threshold multiplier
threshold = 1.5

# Identify outliers
lower_bound = Q1 - (threshold * IQR)
upper_bound = Q3 + (threshold * IQR)

# Data points outside the range [lower_bound, upper_bound] are considered outliers
outliers = ((selected_data < lower_bound) | (selected_data > upper_bound)).any(axis=1)

# Remove outliers
cleaned_data = selected_data[~outliers]

# Display the cleaned data
print(cleaned_data)
```

```python
import statsmodels.api as sm

# Define the dependent variable (target) and independent variables
target_variable = 'popularity'  # You can change this to any other dependent variable
independent_variables = ['budget', 'vote_count', 'vote_average']

# Add constant term to independent variables
X = sm.add_constant(cleaned_data[independent_variables])

# Define the dependent variable
y = cleaned_data[target_variable]

# Fit OLS regression model
model = sm.OLS(y, X).fit()

# Print model summary
print(model.summary())
```

```python
import numpy as np
# Make predictions
y_pred = model.predict(X)

# Calculate Mean Squared Error (MSE)
mse = np.mean((y - y_pred)**2)

# Calculate Root Mean Squared Error (RMSE)
rmse = np.sqrt(mse)

# Calculate Mean Absolute Error (MAE)
mae = np.mean(np.abs(y - y_pred))

# Calculate R-squared (R2 score)
r2 = model.rsquared

print("Mean Squared Error:", mse)
print("Root Mean Squared Error:", rmse)
print("R-squared (R2) Score:", r2)
print("Mean Absolute Error:", mae)
```
